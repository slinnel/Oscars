```python
#SAGS
sagsAll={}
for i in years: 
    print(i)
    url = 'http://www.imdb.com/event/ev0000598/'+ i
    page = urllib2.urlopen(url)
    #pageSplit = page
    soup = BeautifulSoup(page.read(), "html5lib")
    html = u""
    awardName="Outstanding Performance by a Cast in a Motion Picture"
    if(i=="1996"):
        awardName="Outstanding Performance by a Cast"
    if(i=="1997"):
        awardName="Outstanding Performance by a Cast"
    if(i=="1998"):
        awardName="Outstanding Performance by a Cast"
    if(i=="1999"):
        awardName="Outstanding Performance by a Cast"
    if(i=="2000"):
        awardName="Outstanding Performance by a Cast in a Theatrical Motion Picture"
    if(i=="2001"):
        awardName="Outstanding Performance by the Cast of a Theatrical Motion Picture"
    if(i=="2002"):
        awardName="Outstanding Performance by the Cast of a Theatrical Motion Picture"
    if(i=="2003"):
        awardName="Outstanding Performance by the Cast of a Theatrical Motion Picture"  

        
    for tag in soup.find("h2",text=awardName).next_siblings:
        if tag.name == "h2":
            break
        else:
            html += unicode(tag)
    #page = soup.findAll("div",{"class":"award"})
    #awards = page[0].findAll("blockquote")
    awards= BeautifulSoup(html, "html5lib")
    film = awards.findAll("strong")
    thisYear = []
    for j in film: 
        thisYear.append(j.a.text)

    sagsAll[int(i)]=thisYear
all["sags"]=sagsAll
```
